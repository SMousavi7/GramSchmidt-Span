```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 0])
y = np.array([0, 0])
a = np.array([1, 2])
b = np.array([3, -1])

plt.quiver(x, y, a, b, angles = 'xy', scale_units = 'xy', scale = 1)
scaler = np.arange(-5, 5, 0.001)
scaler = scaler.reshape(scaler.shape[0], 1)
np.random.shuffle(scaler)

scaler2 = np.arange(-5, 5, 0.001)
scaler2 = scaler2.reshape(scaler2.shape[0], 1)
np.random.shuffle(scaler2)
print(a[1])

span_a = a[0] * scaler + b[0] * scaler2
span_b = a[1] * scaler + b[1] * scaler2

plt.scatter(span_a, span_b, color = 'b', marker = '*')

plt.xticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
